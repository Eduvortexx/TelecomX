```python
df_dados_normalizados = pd.concat([
    df.drop(columns=['customer', 'phone', 'internet','account']),
    pd.json_normalize(df['customer']),
    pd.json_normalize(df['phone']),
    pd.json_normalize(df['internet']),
    pd.json_normalize(df['account']),
], axis=1)
```

```python
# Dicion√°rios de mapeamento para as tradu√ß√µes
map_churn = {'Yes': 'Sim', 'No': 'N√£o'}
map_gender = {'Female': 'Feminino', 'Male': 'Masculino'}
map_partner = {'Yes': 'Sim', 'No': 'N√£o'}
map_dependents = {'Yes': 'Sim', 'No': 'N√£o'}
map_phone_service = {'Yes': 'Sim', 'No': 'N√£o'}
map_multiple_lines = {'No phone service': 'Sem servi√ßo telef√¥nico', 'No': 'N√£o', 'Yes': 'Sim'}
map_internet_service = {'DSL': 'DSL', 'Fiber optic': 'Fibra √≥tica', 'No': 'N√£o'}
map_online_security = {'No internet service': 'Sem servi√ßo de internet', 'No': 'N√£o', 'Yes': 'Sim'}
map_online_backup = {'No internet service': 'Sem servi√ßo de internet', 'No': 'N√£o', 'Yes': 'Sim'}
map_device_protection = {'No internet service': 'Sem servi√ßo de internet', 'No': 'N√£o', 'Yes': 'Sim'}
map_tech_support = {'No internet service': 'Sem servi√ßo de internet', 'No': 'N√£o', 'Yes': 'Sim'}
map_streaming_tv =  {'No internet service': 'Sem servi√ßo de internet', 'No': 'N√£o', 'Yes': 'Sim'}
map_streaming_movies =  {'No internet service': 'Sem servi√ßo de internet', 'No': 'N√£o', 'Yes': 'Sim'}
map_contract = {'Month-to-month': 'M√™s a m√™s', 'One year': 'Um ano', 'Two year': 'Dois anos'}
map_paperless_billing = {'Yes': 'Sim', 'No': 'N√£o'}
map_payment_method = {'Electronic check': 'Cheque eletr√¥nico', 'Mailed check': 'Cheque por correio', 'Bank transfer (automatic)': 'Transfer√™ncia banc√°ria (autom√°tica)', 'Credit card (automatic)': 'Cart√£o de cr√©dito (autom√°tico)'}

# Aplicar as tradu√ß√µes
df_dados_normalizados['Churn'] = df_dados_normalizados['Churn'].replace(map_churn)
df_dados_normalizados['gender'] = df_dados_normalizados['gender'].replace(map_gender)
df_dados_normalizados['Partner'] = df_dados_normalizados['Partner'].replace(map_partner)
df_dados_normalizados['Dependents'] = df_dados_normalizados['Dependents'].replace(map_dependents)
df_dados_normalizados['PhoneService'] = df_dados_normalizados['PhoneService'].replace(map_phone_service)
df_dados_normalizados['MultipleLines'] = df_dados_normalizados['MultipleLines'].replace(map_multiple_lines)
df_dados_normalizados['InternetService'] = df_dados_normalizados['InternetService'].replace(map_internet_service)
df_dados_normalizados['OnlineSecurity'] = df_dados_normalizados['OnlineSecurity'].replace(map_online_security)
df_dados_normalizados['OnlineBackup'] = df_dados_normalizados['OnlineBackup'].replace(map_online_backup)
df_dados_normalizados['DeviceProtection'] = df_dados_normalizados['DeviceProtection'].replace(map_device_protection)
df_dados_normalizados['TechSupport'] = df_dados_normalizados['TechSupport'].replace(map_tech_support)
df_dados_normalizados['StreamingTV'] = df_dados_normalizados['StreamingTV'].replace(map_streaming_tv)
df_dados_normalizados['StreamingMovies'] = df_dados_normalizados['StreamingMovies'].replace(map_streaming_movies)
df_dados_normalizados['Contract'] = df_dados_normalizados['Contract'].replace(map_contract)
df_dados_normalizados['PaperlessBilling'] = df_dados_normalizados['PaperlessBilling'].replace(map_paperless_billing)
df_dados_normalizados['PaymentMethod'] = df_dados_normalizados['PaymentMethod'].replace(map_payment_method)

# Exibir os valores √∫nicos das colunas traduzidas para verifica√ß√£o
print("\nValores √∫nicos ap√≥s tradu√ß√£o:")
for coluna in ['Churn', 'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']:
    print(f"\nColuna: {coluna}")
    print(df_dados_normalizados[coluna].unique())

# Definir o mapa de renomea√ß√£o
mapa_de_nomes = {
 'customerID': 'ID_Cliente',
    'Churn': 'Cancelou',
    'gender': 'Genero',
    'SeniorCitizen': 'Idoso',
    'Partner': 'Possui_Conjuge',
    'Dependents': 'Possui_Dependentes',
    'tenure': 'Meses_Permanencia',
    'PhoneService': 'Servico_Telefone',
    'MultipleLines': 'Multiplas_Linhas',
    'InternetService': 'Tipo_Internet',
    'OnlineBackup': 'Backup_Online',
    'DeviceProtection': 'Protecao_Dispositivo',
    'TechSupport': 'Suporte_Tecnico',
    'StreamingTV': 'Streaming_TV',
    'StreamingMovies': 'Streaming_Filmes',
    'Contract': 'Tipo_Contrato',
    'PaperlessBilling': 'Fatura_Digital',
    'PaymentMethod': 'Metodo_Pagamento',
    'Charges.Monthly': 'Cobranca_Mensal',
    'Charges.Total': 'Cobranca_Total'

}

# Renomear imediatamente
df_final = df_dados_normalizados.rename(columns=mapa_de_nomes)

# 4. Verificar o resultado
df_final.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Carregamento e Prepara√ß√£o dos Dados ---

# Configura√ß√µes de estilo para os gr√°ficos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 7)
plt.rcParams['font.size'] = 12

# Limpeza final: Remover linhas onde a coluna 'Cancelou' est√° vazia
# Primeiro, substitu√≠mos os valores vazios ('') por um marcador de aus√™ncia (NA)
df_final['Cancelou'].replace('', pd.NA, inplace=True)
# Em seguida, removemos todas as linhas que cont√™m esse marcador na coluna 'Cancelou'
df_final.dropna(subset=['Cancelou'], inplace=True)

# Para evitar avisos de c√≥pia, criamos uma c√≥pia expl√≠cita do DataFrame filtrado
df = df_final.copy()

# Criar uma coluna de r√≥tulo para 'Idoso' para melhorar a visualiza√ß√£o nos gr√°ficos
df['Cliente_Idoso'] = df['Idoso'].map({0: 'N√£o', 1: 'Sim'})

print(f"Dados carregados e limpos. Total de {len(df)} registros para an√°lise.")
print("\nDistribui√ß√£o de Cancelamentos:")
print(df['Cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```

```python
# Gr√°fico 1: Distribui√ß√£o Geral de Cancelamentos
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Cancelou', data=df_final, palette='pastel', order=['N√£o', 'Sim'])
ax.set_title('Distribui√ß√£o Geral de Cancelamentos', fontsize=16)
ax.set_xlabel('Cliente Cancelou o Servi√ßo?', fontsize=12)
ax.set_ylabel('N√∫mero de Clientes', fontsize=12)

# Adicionar anota√ß√µes de porcentagem no gr√°fico
total = len(df_final)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=12, color='black')


plt.show()
```

```python
# Gr√°fico 2 : Propor√ß√£o de Cancelamento por Tipo de Contrato

# Passo 1: Calcular a propor√ß√£o de cancelamento para cada tipo de contrato
# Usamos a fun√ß√£o crosstab para criar uma tabela de conting√™ncia
# e normalizamos por linha ('index') para obter as porcentagens.
contract_churn_proportion = pd.crosstab(df['Tipo_Contrato'], df['Cancelou'], normalize='index')

# Reordenar as colunas para que 'Sim' (cancelamento) apare√ßa primeiro e com cor de destaque
contract_churn_proportion = contract_churn_proportion[['Sim', 'N√£o']]

print("Propor√ß√£o de Cancelamento por Tipo de Contrato:")
print(contract_churn_proportion.applymap('{:.2%}'.format))

# Passo 2: Gerar o gr√°fico de barras empilhadas 100%
ax = contract_churn_proportion.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 8),
    color=['#d62728', '#1f77b4'],  # Vermelho para 'Sim' (alerta), Azul para 'N√£o'
    edgecolor='white',
    width=0.7
)

# Passo 3: Adicionar anota√ß√µes e melhorar a legibilidade do gr√°fico
plt.title('Propor√ß√£o de Cancelamento por Tipo de Contrato', fontsize=16, pad=20)
plt.xlabel('Tipo de Contrato', fontsize=12, labelpad=15)
plt.ylabel('Porcentagem de Clientes', fontsize=12)
plt.xticks(rotation=0, fontsize=11) # Mant√©m os r√≥tulos do eixo X na horizontal

# Formatar o eixo Y para mostrar o s√≠mbolo de porcentagem
ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
plt.legend(title='Cancelou?', bbox_to_anchor=(1.02, 1), loc='upper left')

# Adicionar os r√≥tulos de porcentagem dentro de cada barra para clareza
for container in ax.containers:
    # Formata o r√≥tulo para mostrar a porcentagem, mas s√≥ se for maior que 5%
    labels = [f'{v.get_height():.1%}' if v.get_height() > 0.05 else '' for v in container]
    ax.bar_label(container, labels=labels, label_type='center', color='white', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Gr√°fico 3: Rela√ß√£o entre Cobran√ßa Mensal e Cancelamento
plt.figure(figsize=(12, 7))
sns.boxplot(x='Cancelou', y='Cobranca_Mensal', data=df, palette='coolwarm')
plt.title('Distribui√ß√£o da Cobran√ßa Mensal por Status de Cancelamento', fontsize=16)
plt.xlabel('Cliente Cancelou o Servi√ßo?', fontsize=12)
plt.ylabel('Cobran√ßa Mensal', fontsize=12)
plt.show()
```

```python
# Gr√°fico 4: Rela√ß√£o entre Meses de Perman√™ncia e Cancelamento
plt.figure(figsize=(12, 7))
sns.histplot(data=df, x='Meses_Permanencia', hue='Cancelou', multiple='stack', palette='magma', bins=30, kde=True)
plt.title('Influ√™ncia dos Meses de Perman√™ncia no Cancelamento', fontsize=16)
plt.xlabel('Meses de Perman√™ncia na Empresa', fontsize=12)
plt.ylabel('N√∫mero de Clientes', fontsize=12)
plt.legend(title='Cancelou?', labels=['Sim', 'N√£o'])
plt.show()
```

```python
# Gr√°fico 5: Propor√ß√£o de Cancelamento por Fatores Demogr√°ficos

# Lista das colunas demogr√°ficas que queremos analisar
demographic_columns = ['Genero', 'Cliente_Idoso', 'Possui_Conjuge', 'Possui_Dependentes']

# Criando a figura com subplots
# O n√∫mero de linhas se ajustar√° com base no n√∫mero de colunas a serem plotadas
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))
fig.suptitle('Propor√ß√£o de Cancelamento por Fatores Demogr√°ficos', fontsize=20, y=1.02)

# "Flatten" o array de eixos para facilitar a itera√ß√£o
axes = axes.flatten()

# Loop para criar um gr√°fico para cada coluna demogr√°fica
for i, col in enumerate(demographic_columns):
    ax = axes[i]

    # Calcula a propor√ß√£o de cancelamento para a coluna atual
    proportion = pd.crosstab(df[col], df['Cancelou'], normalize='index')
    proportion = proportion[['Sim', 'N√£o']] # Garante a ordem e cores consistentes

    # Plota o gr√°fico de barras empilhadas 100%
    proportion.plot(
        kind='bar',
        stacked=True,
        ax=ax,
        color=['#d62728', '#1f77b4'], # Vermelho para 'Sim' (alerta), Azul para 'N√£o'
        width=0.6,
        edgecolor='white'
    )

    # Adiciona os r√≥tulos de porcentagem dentro das barras
    for container in ax.containers:
        labels = [f'{v.get_height():.1%}' if v.get_height() > 0.05 else '' for v in container]
        ax.bar_label(container, labels=labels, label_type='center', color='white', fontsize=11, fontweight='bold')

    # Configura√ß√µes de est√©tica do gr√°fico
    ax.set_title(f'Cancelamento por {col.replace("_", " ")}', fontsize=14)
    ax.set_ylabel('Porcentagem de Clientes', fontsize=12)
    ax.set_xlabel('')
    ax.tick_params(axis='x', rotation=0, labelsize=12)
    ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
    ax.get_legend().set_title('Cancelou?')

# Remove eixos extras se o n√∫mero de gr√°ficos for √≠mpar
if len(demographic_columns) < len(axes):
    for i in range(len(demographic_columns), len(axes)):
        fig.delaxes(axes[i])

plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```

#üìÑRelatorio Final

üìä An√°lise de Cancelamentos na TelecomX
Introdu√ß√£o: Compreendendo o Churn
Esta an√°lise tem como foco principal identificar os fatores que contribuem para o desligamento de clientes da TelecomX ‚Äî um fen√¥meno conhecido como churn. Saber o que impulsiona esse comportamento √© vital para a empresa, tanto do ponto de vista financeiro quanto estrat√©gico. Reter clientes √© mais econ√¥mico do que conquistar novos, e altas taxas de churn podem prejudicar a reputa√ß√£o da marca a longo prazo.

Prepara√ß√£o dos Dados: Do Bruto ao Estruturado
Utilizamos como base um arquivo chamado TelecomX_data. As etapas de prepara√ß√£o inclu√≠ram:

Importa√ß√£o dos Dados: Os registros foram carregados em um DataFrame utilizando a biblioteca pandas, o que facilitou a organiza√ß√£o e manipula√ß√£o das informa√ß√µes.

Limpeza: Removemos entradas com informa√ß√µes ausentes na vari√°vel que indica se o cliente cancelou ou n√£o.

Transforma√ß√£o: A vari√°vel "Cancelou", originalmente textual, foi convertida para valores num√©ricos ‚Äî 1 representando cancelamento e 0, perman√™ncia ‚Äî permitindo an√°lises quantitativas e uso em modelos preditivos.

Ap√≥s esse processo, obtivemos um conjunto de 7.032 clientes, dos quais 26,6% encerraram seus contratos com a empresa.

Explorando os Dados: O Que Est√° por Tr√°s do Cancelamento?
A partir da an√°lise das principais vari√°veis demogr√°ficas e comportamentais, destacam-se os seguintes insights:

üîπ G√™nero
A propor√ß√£o de cancelamento entre homens (26,2%) e mulheres (26,9%) √© praticamente a mesma. Isso indica que o sexo do cliente n√£o exerce influ√™ncia significativa sobre o churn.

üîπ Idade Avan√ßada
Clientes classificados como idosos apresentam um √≠ndice de cancelamento substancialmente superior (41,7%) em compara√ß√£o aos n√£o idosos (23,6%). A idade, portanto, √© um forte preditor de risco.

üîπ Estado Civil
A presen√ßa de um c√¥njuge se mostra relevante: clientes solteiros ou sem parceiro cancelam em uma taxa de 33,0%, contra apenas 19,7% daqueles que vivem com algu√©m. Isso sugere que la√ßos familiares podem contribuir para maior reten√ß√£o.

üîπ Dependentes
O comportamento √© similar ao anterior: clientes sem dependentes t√™m taxa de churn de 31,3%, enquanto aqueles com dependentes apresentam apenas 15,5%. A responsabilidade familiar parece associada √† fidelidade ao servi√ßo.

Principais Determinantes do Churn
Al√©m dos fatores demogr√°ficos, as vari√°veis contratuais e financeiras tamb√©m demonstram grande relev√¢ncia:

Tipo de Contrato: Contratos mensais s√£o os que apresentam maior vulnerabilidade ao churn. A liberdade de cancelamento sem fideliza√ß√£o pode estar levando muitos clientes a desistirem do servi√ßo.

Tempo de Perman√™ncia (Tenure): A maior parte dos cancelamentos ocorre nos primeiros meses. Clientes com maior tempo de v√≠nculo tendem a permanecer, indicando uma rela√ß√£o direta entre tempo de casa e lealdade.

Valor da Fatura Mensal: Os que encerram o contrato geralmente possuem faturas mais elevadas, sugerindo uma poss√≠vel insatisfa√ß√£o com o custo-benef√≠cio percebido.

Forma de Pagamento: Aqueles que utilizam cheque eletr√¥nico como forma de pagamento t√™m uma taxa de churn significativamente superior. Isso pode apontar para uma instabilidade nesse perfil ou uma limita√ß√£o associada a esse m√©todo.

Perfil de Alto Risco
Com base nos padr√µes observados, podemos descrever o cliente com maior probabilidade de cancelamento como algu√©m que:

Tem contrato mensal;

√â novo na empresa;

Paga valores altos nas faturas;

Utiliza cheque eletr√¥nico;

Possui idade avan√ßada;

N√£o tem dependentes nem c√¥njuge;

N√£o utiliza servi√ßos adicionais como Suporte T√©cnico ou Seguran√ßa Online.

Estrat√©gias Sugeridas para Redu√ß√£o do Churn
A partir dessas descobertas, algumas a√ß√µes pr√°ticas podem ser implementadas para reduzir a evas√£o de clientes:

Incentivar Contratos de Longo Prazo: Promover benef√≠cios para quem migra de contratos mensais para anuais ou bienais pode aumentar a reten√ß√£o.

Apoiar Novos Clientes: Criar uma jornada de onboarding mais eficiente para os primeiros seis meses pode ser decisivo na fideliza√ß√£o inicial.

Monitorar Contas Altas: Identificar clientes com faturas elevadas e oferecer planos personalizados ou descontos pode diminuir a percep√ß√£o negativa sobre o custo.

Reavaliar o Cheque Eletr√¥nico: Investigar a raz√£o por tr√°s da alta taxa de cancelamento nesse grupo e incentivar alternativas como d√©bito autom√°tico ou cart√£o de cr√©dito.

Aten√ß√£o Especial aos Idosos: Melhorar a comunica√ß√£o, o suporte t√©cnico e a usabilidade dos servi√ßos para esse p√∫blico pode ajudar a reduzir o churn.

Considera√ß√µes Finais
O cancelamento de servi√ßos n√£o acontece ao acaso ‚Äî ele segue padr√µes que podem ser medidos, compreendidos e combatidos. A an√°lise revela que a TelecomX tem condi√ß√µes de intervir estrategicamente para evitar perdas, fortalecendo o relacionamento com o cliente e otimizando seus resultados de forma sustent√°vel.
