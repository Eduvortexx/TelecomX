```python
df_dados_normalizados = pd.concat([
    df.drop(columns=['customer', 'phone', 'internet','account']),
    pd.json_normalize(df['customer']),
    pd.json_normalize(df['phone']),
    pd.json_normalize(df['internet']),
    pd.json_normalize(df['account']),
], axis=1)
```

```python
# DicionÃ¡rios de mapeamento para as traduÃ§Ãµes
map_churn = {'Yes': 'Sim', 'No': 'NÃ£o'}
map_gender = {'Female': 'Feminino', 'Male': 'Masculino'}
map_partner = {'Yes': 'Sim', 'No': 'NÃ£o'}
map_dependents = {'Yes': 'Sim', 'No': 'NÃ£o'}
map_phone_service = {'Yes': 'Sim', 'No': 'NÃ£o'}
map_multiple_lines = {'No phone service': 'Sem serviÃ§o telefÃ´nico', 'No': 'NÃ£o', 'Yes': 'Sim'}
map_internet_service = {'DSL': 'DSL', 'Fiber optic': 'Fibra Ã³tica', 'No': 'NÃ£o'}
map_online_security = {'No internet service': 'Sem serviÃ§o de internet', 'No': 'NÃ£o', 'Yes': 'Sim'}
map_online_backup = {'No internet service': 'Sem serviÃ§o de internet', 'No': 'NÃ£o', 'Yes': 'Sim'}
map_device_protection = {'No internet service': 'Sem serviÃ§o de internet', 'No': 'NÃ£o', 'Yes': 'Sim'}
map_tech_support = {'No internet service': 'Sem serviÃ§o de internet', 'No': 'NÃ£o', 'Yes': 'Sim'}
map_streaming_tv =  {'No internet service': 'Sem serviÃ§o de internet', 'No': 'NÃ£o', 'Yes': 'Sim'}
map_streaming_movies =  {'No internet service': 'Sem serviÃ§o de internet', 'No': 'NÃ£o', 'Yes': 'Sim'}
map_contract = {'Month-to-month': 'MÃªs a mÃªs', 'One year': 'Um ano', 'Two year': 'Dois anos'}
map_paperless_billing = {'Yes': 'Sim', 'No': 'NÃ£o'}
map_payment_method = {'Electronic check': 'Cheque eletrÃ´nico', 'Mailed check': 'Cheque por correio', 'Bank transfer (automatic)': 'TransferÃªncia bancÃ¡ria (automÃ¡tica)', 'Credit card (automatic)': 'CartÃ£o de crÃ©dito (automÃ¡tico)'}

# Aplicar as traduÃ§Ãµes
df_dados_normalizados['Churn'] = df_dados_normalizados['Churn'].replace(map_churn)
df_dados_normalizados['gender'] = df_dados_normalizados['gender'].replace(map_gender)
df_dados_normalizados['Partner'] = df_dados_normalizados['Partner'].replace(map_partner)
df_dados_normalizados['Dependents'] = df_dados_normalizados['Dependents'].replace(map_dependents)
df_dados_normalizados['PhoneService'] = df_dados_normalizados['PhoneService'].replace(map_phone_service)
df_dados_normalizados['MultipleLines'] = df_dados_normalizados['MultipleLines'].replace(map_multiple_lines)
df_dados_normalizados['InternetService'] = df_dados_normalizados['InternetService'].replace(map_internet_service)
df_dados_normalizados['OnlineSecurity'] = df_dados_normalizados['OnlineSecurity'].replace(map_online_security)
df_dados_normalizados['OnlineBackup'] = df_dados_normalizados['OnlineBackup'].replace(map_online_backup)
df_dados_normalizados['DeviceProtection'] = df_dados_normalizados['DeviceProtection'].replace(map_device_protection)
df_dados_normalizados['TechSupport'] = df_dados_normalizados['TechSupport'].replace(map_tech_support)
df_dados_normalizados['StreamingTV'] = df_dados_normalizados['StreamingTV'].replace(map_streaming_tv)
df_dados_normalizados['StreamingMovies'] = df_dados_normalizados['StreamingMovies'].replace(map_streaming_movies)
df_dados_normalizados['Contract'] = df_dados_normalizados['Contract'].replace(map_contract)
df_dados_normalizados['PaperlessBilling'] = df_dados_normalizados['PaperlessBilling'].replace(map_paperless_billing)
df_dados_normalizados['PaymentMethod'] = df_dados_normalizados['PaymentMethod'].replace(map_payment_method)

# Exibir os valores Ãºnicos das colunas traduzidas para verificaÃ§Ã£o
print("\nValores Ãºnicos apÃ³s traduÃ§Ã£o:")
for coluna in ['Churn', 'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']:
    print(f"\nColuna: {coluna}")
    print(df_dados_normalizados[coluna].unique())

# Definir o mapa de renomeaÃ§Ã£o
mapa_de_nomes = {
 'customerID': 'ID_Cliente',
    'Churn': 'Cancelou',
    'gender': 'Genero',
    'SeniorCitizen': 'Idoso',
    'Partner': 'Possui_Conjuge',
    'Dependents': 'Possui_Dependentes',
    'tenure': 'Meses_Permanencia',
    'PhoneService': 'Servico_Telefone',
    'MultipleLines': 'Multiplas_Linhas',
    'InternetService': 'Tipo_Internet',
    'OnlineBackup': 'Backup_Online',
    'DeviceProtection': 'Protecao_Dispositivo',
    'TechSupport': 'Suporte_Tecnico',
    'StreamingTV': 'Streaming_TV',
    'StreamingMovies': 'Streaming_Filmes',
    'Contract': 'Tipo_Contrato',
    'PaperlessBilling': 'Fatura_Digital',
    'PaymentMethod': 'Metodo_Pagamento',
    'Charges.Monthly': 'Cobranca_Mensal',
    'Charges.Total': 'Cobranca_Total'

}

# Renomear imediatamente
df_final = df_dados_normalizados.rename(columns=mapa_de_nomes)

# 4. Verificar o resultado
df_final.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Carregamento e PreparaÃ§Ã£o dos Dados ---

# ConfiguraÃ§Ãµes de estilo para os grÃ¡ficos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 7)
plt.rcParams['font.size'] = 12

# Limpeza final: Remover linhas onde a coluna 'Cancelou' estÃ¡ vazia
# Primeiro, substituÃ­mos os valores vazios ('') por um marcador de ausÃªncia (NA)
df_final['Cancelou'].replace('', pd.NA, inplace=True)
# Em seguida, removemos todas as linhas que contÃªm esse marcador na coluna 'Cancelou'
df_final.dropna(subset=['Cancelou'], inplace=True)

# Para evitar avisos de cÃ³pia, criamos uma cÃ³pia explÃ­cita do DataFrame filtrado
df = df_final.copy()

# Criar uma coluna de rÃ³tulo para 'Idoso' para melhorar a visualizaÃ§Ã£o nos grÃ¡ficos
df['Cliente_Idoso'] = df['Idoso'].map({0: 'NÃ£o', 1: 'Sim'})

print(f"Dados carregados e limpos. Total de {len(df)} registros para anÃ¡lise.")
print("\nDistribuiÃ§Ã£o de Cancelamentos:")
print(df['Cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```

```python
# GrÃ¡fico 1: DistribuiÃ§Ã£o Geral de Cancelamentos
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Cancelou', data=df_final, palette='pastel', order=['NÃ£o', 'Sim'])
ax.set_title('DistribuiÃ§Ã£o Geral de Cancelamentos', fontsize=16)
ax.set_xlabel('Cliente Cancelou o ServiÃ§o?', fontsize=12)
ax.set_ylabel('NÃºmero de Clientes', fontsize=12)

# Adicionar anotaÃ§Ãµes de porcentagem no grÃ¡fico
total = len(df_final)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=12, color='black')


plt.show()
```

```python
# GrÃ¡fico 2 : ProporÃ§Ã£o de Cancelamento por Tipo de Contrato

# Passo 1: Calcular a proporÃ§Ã£o de cancelamento para cada tipo de contrato
# Usamos a funÃ§Ã£o crosstab para criar uma tabela de contingÃªncia
# e normalizamos por linha ('index') para obter as porcentagens.
contract_churn_proportion = pd.crosstab(df['Tipo_Contrato'], df['Cancelou'], normalize='index')

# Reordenar as colunas para que 'Sim' (cancelamento) apareÃ§a primeiro e com cor de destaque
contract_churn_proportion = contract_churn_proportion[['Sim', 'NÃ£o']]

print("ProporÃ§Ã£o de Cancelamento por Tipo de Contrato:")
print(contract_churn_proportion.applymap('{:.2%}'.format))

# Passo 2: Gerar o grÃ¡fico de barras empilhadas 100%
ax = contract_churn_proportion.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 8),
    color=['#d62728', '#1f77b4'],  # Vermelho para 'Sim' (alerta), Azul para 'NÃ£o'
    edgecolor='white',
    width=0.7
)

# Passo 3: Adicionar anotaÃ§Ãµes e melhorar a legibilidade do grÃ¡fico
plt.title('ProporÃ§Ã£o de Cancelamento por Tipo de Contrato', fontsize=16, pad=20)
plt.xlabel('Tipo de Contrato', fontsize=12, labelpad=15)
plt.ylabel('Porcentagem de Clientes', fontsize=12)
plt.xticks(rotation=0, fontsize=11) # MantÃ©m os rÃ³tulos do eixo X na horizontal

# Formatar o eixo Y para mostrar o sÃ­mbolo de porcentagem
ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
plt.legend(title='Cancelou?', bbox_to_anchor=(1.02, 1), loc='upper left')

# Adicionar os rÃ³tulos de porcentagem dentro de cada barra para clareza
for container in ax.containers:
    # Formata o rÃ³tulo para mostrar a porcentagem, mas sÃ³ se for maior que 5%
    labels = [f'{v.get_height():.1%}' if v.get_height() > 0.05 else '' for v in container]
    ax.bar_label(container, labels=labels, label_type='center', color='white', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# GrÃ¡fico 3: RelaÃ§Ã£o entre CobranÃ§a Mensal e Cancelamento
plt.figure(figsize=(12, 7))
sns.boxplot(x='Cancelou', y='Cobranca_Mensal', data=df, palette='coolwarm')
plt.title('DistribuiÃ§Ã£o da CobranÃ§a Mensal por Status de Cancelamento', fontsize=16)
plt.xlabel('Cliente Cancelou o ServiÃ§o?', fontsize=12)
plt.ylabel('CobranÃ§a Mensal', fontsize=12)
plt.show()
```

```python
# GrÃ¡fico 4: RelaÃ§Ã£o entre Meses de PermanÃªncia e Cancelamento
plt.figure(figsize=(12, 7))
sns.histplot(data=df, x='Meses_Permanencia', hue='Cancelou', multiple='stack', palette='magma', bins=30, kde=True)
plt.title('InfluÃªncia dos Meses de PermanÃªncia no Cancelamento', fontsize=16)
plt.xlabel('Meses de PermanÃªncia na Empresa', fontsize=12)
plt.ylabel('NÃºmero de Clientes', fontsize=12)
plt.legend(title='Cancelou?', labels=['Sim', 'NÃ£o'])
plt.show()
```

```python
# GrÃ¡fico 5: ProporÃ§Ã£o de Cancelamento por Fatores DemogrÃ¡ficos

# Lista das colunas demogrÃ¡ficas que queremos analisar
demographic_columns = ['Genero', 'Cliente_Idoso', 'Possui_Conjuge', 'Possui_Dependentes']

# Criando a figura com subplots
# O nÃºmero de linhas se ajustarÃ¡ com base no nÃºmero de colunas a serem plotadas
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))
fig.suptitle('ProporÃ§Ã£o de Cancelamento por Fatores DemogrÃ¡ficos', fontsize=20, y=1.02)

# "Flatten" o array de eixos para facilitar a iteraÃ§Ã£o
axes = axes.flatten()

# Loop para criar um grÃ¡fico para cada coluna demogrÃ¡fica
for i, col in enumerate(demographic_columns):
    ax = axes[i]

    # Calcula a proporÃ§Ã£o de cancelamento para a coluna atual
    proportion = pd.crosstab(df[col], df['Cancelou'], normalize='index')
    proportion = proportion[['Sim', 'NÃ£o']] # Garante a ordem e cores consistentes

    # Plota o grÃ¡fico de barras empilhadas 100%
    proportion.plot(
        kind='bar',
        stacked=True,
        ax=ax,
        color=['#d62728', '#1f77b4'], # Vermelho para 'Sim' (alerta), Azul para 'NÃ£o'
        width=0.6,
        edgecolor='white'
    )

    # Adiciona os rÃ³tulos de porcentagem dentro das barras
    for container in ax.containers:
        labels = [f'{v.get_height():.1%}' if v.get_height() > 0.05 else '' for v in container]
        ax.bar_label(container, labels=labels, label_type='center', color='white', fontsize=11, fontweight='bold')

    # ConfiguraÃ§Ãµes de estÃ©tica do grÃ¡fico
    ax.set_title(f'Cancelamento por {col.replace("_", " ")}', fontsize=14)
    ax.set_ylabel('Porcentagem de Clientes', fontsize=12)
    ax.set_xlabel('')
    ax.tick_params(axis='x', rotation=0, labelsize=12)
    ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
    ax.get_legend().set_title('Cancelou?')

# Remove eixos extras se o nÃºmero de grÃ¡ficos for Ã­mpar
if len(demographic_columns) < len(axes):
    for i in range(len(demographic_columns), len(axes)):
        fig.delaxes(axes[i])

plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```

ðŸ“Š AnÃ¡lise de Cancelamentos na TelecomX
IntroduÃ§Ã£o: Compreendendo o Churn
Esta anÃ¡lise tem como foco principal identificar os fatores que contribuem para o desligamento de clientes da TelecomX â€” um fenÃ´meno conhecido como churn. Saber o que impulsiona esse comportamento Ã© vital para a empresa, tanto do ponto de vista financeiro quanto estratÃ©gico. Reter clientes Ã© mais econÃ´mico do que conquistar novos, e altas taxas de churn podem prejudicar a reputaÃ§Ã£o da marca a longo prazo.

PreparaÃ§Ã£o dos Dados: Do Bruto ao Estruturado
Utilizamos como base um arquivo chamado TelecomX_data. As etapas de preparaÃ§Ã£o incluÃ­ram:

ImportaÃ§Ã£o dos Dados: Os registros foram carregados em um DataFrame utilizando a biblioteca pandas, o que facilitou a organizaÃ§Ã£o e manipulaÃ§Ã£o das informaÃ§Ãµes.

Limpeza: Removemos entradas com informaÃ§Ãµes ausentes na variÃ¡vel que indica se o cliente cancelou ou nÃ£o.

TransformaÃ§Ã£o: A variÃ¡vel "Cancelou", originalmente textual, foi convertida para valores numÃ©ricos â€” 1 representando cancelamento e 0, permanÃªncia â€” permitindo anÃ¡lises quantitativas e uso em modelos preditivos.

ApÃ³s esse processo, obtivemos um conjunto de 7.032 clientes, dos quais 26,6% encerraram seus contratos com a empresa.

Explorando os Dados: O Que EstÃ¡ por TrÃ¡s do Cancelamento?
A partir da anÃ¡lise das principais variÃ¡veis demogrÃ¡ficas e comportamentais, destacam-se os seguintes insights:

ðŸ”¹ GÃªnero
A proporÃ§Ã£o de cancelamento entre homens (26,2%) e mulheres (26,9%) Ã© praticamente a mesma. Isso indica que o sexo do cliente nÃ£o exerce influÃªncia significativa sobre o churn.

ðŸ”¹ Idade AvanÃ§ada
Clientes classificados como idosos apresentam um Ã­ndice de cancelamento substancialmente superior (41,7%) em comparaÃ§Ã£o aos nÃ£o idosos (23,6%). A idade, portanto, Ã© um forte preditor de risco.

ðŸ”¹ Estado Civil
A presenÃ§a de um cÃ´njuge se mostra relevante: clientes solteiros ou sem parceiro cancelam em uma taxa de 33,0%, contra apenas 19,7% daqueles que vivem com alguÃ©m. Isso sugere que laÃ§os familiares podem contribuir para maior retenÃ§Ã£o.

ðŸ”¹ Dependentes
O comportamento Ã© similar ao anterior: clientes sem dependentes tÃªm taxa de churn de 31,3%, enquanto aqueles com dependentes apresentam apenas 15,5%. A responsabilidade familiar parece associada Ã  fidelidade ao serviÃ§o.

Principais Determinantes do Churn
AlÃ©m dos fatores demogrÃ¡ficos, as variÃ¡veis contratuais e financeiras tambÃ©m demonstram grande relevÃ¢ncia:

Tipo de Contrato: Contratos mensais sÃ£o os que apresentam maior vulnerabilidade ao churn. A liberdade de cancelamento sem fidelizaÃ§Ã£o pode estar levando muitos clientes a desistirem do serviÃ§o.

Tempo de PermanÃªncia (Tenure): A maior parte dos cancelamentos ocorre nos primeiros meses. Clientes com maior tempo de vÃ­nculo tendem a permanecer, indicando uma relaÃ§Ã£o direta entre tempo de casa e lealdade.

Valor da Fatura Mensal: Os que encerram o contrato geralmente possuem faturas mais elevadas, sugerindo uma possÃ­vel insatisfaÃ§Ã£o com o custo-benefÃ­cio percebido.

Forma de Pagamento: Aqueles que utilizam cheque eletrÃ´nico como forma de pagamento tÃªm uma taxa de churn significativamente superior. Isso pode apontar para uma instabilidade nesse perfil ou uma limitaÃ§Ã£o associada a esse mÃ©todo.

Perfil de Alto Risco
Com base nos padrÃµes observados, podemos descrever o cliente com maior probabilidade de cancelamento como alguÃ©m que:

Tem contrato mensal;

Ã‰ novo na empresa;

Paga valores altos nas faturas;

Utiliza cheque eletrÃ´nico;

Possui idade avanÃ§ada;

NÃ£o tem dependentes nem cÃ´njuge;

NÃ£o utiliza serviÃ§os adicionais como Suporte TÃ©cnico ou SeguranÃ§a Online.

EstratÃ©gias Sugeridas para ReduÃ§Ã£o do Churn
A partir dessas descobertas, algumas aÃ§Ãµes prÃ¡ticas podem ser implementadas para reduzir a evasÃ£o de clientes:

Incentivar Contratos de Longo Prazo: Promover benefÃ­cios para quem migra de contratos mensais para anuais ou bienais pode aumentar a retenÃ§Ã£o.

Apoiar Novos Clientes: Criar uma jornada de onboarding mais eficiente para os primeiros seis meses pode ser decisivo na fidelizaÃ§Ã£o inicial.

Monitorar Contas Altas: Identificar clientes com faturas elevadas e oferecer planos personalizados ou descontos pode diminuir a percepÃ§Ã£o negativa sobre o custo.

Reavaliar o Cheque EletrÃ´nico: Investigar a razÃ£o por trÃ¡s da alta taxa de cancelamento nesse grupo e incentivar alternativas como dÃ©bito automÃ¡tico ou cartÃ£o de crÃ©dito.

AtenÃ§Ã£o Especial aos Idosos: Melhorar a comunicaÃ§Ã£o, o suporte tÃ©cnico e a usabilidade dos serviÃ§os para esse pÃºblico pode ajudar a reduzir o churn.

ConsideraÃ§Ãµes Finais
O cancelamento de serviÃ§os nÃ£o acontece ao acaso â€” ele segue padrÃµes que podem ser medidos, compreendidos e combatidos. A anÃ¡lise revela que a TelecomX tem condiÃ§Ãµes de intervir estrategicamente para evitar perdas, fortalecendo o relacionamento com o cliente e otimizando seus resultados de forma sustentÃ¡vel.
